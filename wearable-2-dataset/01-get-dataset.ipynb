```python
from pymongo import MongoClient
import os
from pymongo.errors import DuplicateKeyError
```

```python
def connect_mongodb(uri, db_name, collection_name):
    # Connect to MongoDB 
    client = MongoClient(uri)
    
    # Select the database and collection
    db = client[db_name]
    collection = db[collection_name]
    
    return collection
def find_documents_with_privacy_compliance(collection):
    """
    Truy vấn tất cả các document trong MongoDB có chứa trường 'privacy-compliance'.

    Args:
        collection: MongoDB collection đã kết nối.

    Returns:
        List các document có chứa trường 'privacy-compliance'.
    """
    # Tìm tất cả các document có chứa key 'privacy-compliance'
    query = {"privacy-compliance": {"$exists": True}}
    results = list(collection.find(query))
    return results
def insert_data_with_existing_id(collection, data):
    """
    Insert a document or list of documents into MongoDB using existing _id values.
    Will raise DuplicateKeyError if a document with the same _id already exists.
    
    Parameters:
    - collection: MongoDB collection object (from connect_mongodb)
    - data: dict or list of dicts, each must include an '_id' key
    """
    try:
        if isinstance(data, list):
            result = collection.insert_many(data, ordered=False)
            print(f"Inserted {len(result.inserted_ids)} documents.")
        elif isinstance(data, dict):
            result = collection.insert_one(data)
            print(f"Inserted document with _id: {result.inserted_id}")
        else:
            raise ValueError("Data must be a dict or list of dicts.")
    except DuplicateKeyError as e:
        print(f"Duplicate key error: {e}")
```

```python
# server
mongoDB_uri = 'mongodb://localhost:27017'
# database-1
mongoDB_database_1 = 'wearable-project' 
mongoDB_collection_1a = 'wearable-app-02-04-2025'
mongoDB_collection_1b = 'wearable-standalone-02-04-2025'
# database-2
mongoDB_database_2 = 'wearable-project-2' 
mongoDB_collection_2 = 'dataset'
```

```python
# Connect to the MongoDB collection
connection_1a = connect_mongodb(mongoDB_uri, mongoDB_database_1, mongoDB_collection_1a)
connection_1b = connect_mongodb(mongoDB_uri, mongoDB_database_1, mongoDB_collection_1b)
connection_2 = connect_mongodb(mongoDB_uri, mongoDB_database_2, mongoDB_collection_2)
```

```python
results_1 = find_documents_with_privacy_compliance(connection_1a)
print(len(results_1))
```

```python
for i in range(len(results_1)):
    print("------------------- Loop-"+str(i)+"-------------------")
    data = results_1[i]
    insert_data_with_existing_id(connection_2, data)
    # break
```

```python
results_2 = find_documents_with_privacy_compliance(connection_1b)
print(len(results_2))
```

```python
for j in range(len(results_2)):
    print("------------------- Loop-"+str(i)+"-------------------")
    data = results_2[j]
    insert_data_with_existing_id(connection_2, data)
    # break
```
