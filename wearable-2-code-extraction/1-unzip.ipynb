```python
import os
import zipfile
import time
```

```python
source_directory = r"E:\wearable-project-decompile-source\source-code-standalone-zip"
destination_directory = r"E:\wearable-project-decompile-source\source-code-standalone-unzip"
```

```python
def get_zip_files(source_directory):
    """Get a list of all .zip files in the specified directory without the .zip extension."""
    zip_files = []
    
    for filename in os.listdir(source_directory):
        # Check if the file ends with .zip
        if filename.endswith(".zip"):
            # Remove the .zip extension
            zip_files.append(os.path.splitext(filename)[0])
    
    return zip_files
zip_files = get_zip_files(source_directory)
#print(zip_files)
print(len(zip_files))
```

```python
def get_directory_names(path):
    """Get a list of all directory names under the specified path."""
    directories = []
    
    # List all entries in the given path
    for entry in os.listdir(path):
        # Construct full path and check if it's a directory
        full_path = os.path.join(path, entry)
        if os.path.isdir(full_path):
            directories.append(entry)
    
    return directories

# Usage
path = r"E:\wearable-project-decompile-source\source-code-unzip"
directory_names = get_directory_names(source_directory)
print(len(directory_names))
```

```python
def remove_zip_files(directory, zip_file_names):
    for file_name in zip_file_names:
        # Construct the full path to each zip file by appending .zip
        zip_file_path = os.path.join(directory, f"{file_name}.zip")
        
        # Check if the file exists before attempting to remove
        if os.path.isfile(zip_file_path):
            try:
                os.remove(zip_file_path)
                print(f"Removed: {zip_file_path}")
            except Exception as e:
                print(f"Error removing {zip_file_path}: {e}")
        else:
            print(f"File does not exist: {zip_file_path}")
remove_zip_files(source_directory, directory_names)
```

```python
# Log file path
log_file_path = "error_log.txt"

# Clear previous log file if exists
with open(log_file_path, "w") as log_file:
    log_file.write("Error Log for Zip Extraction:\n")
```

```python
# Loop through each file in the source directory
for i in range(len(zip_files)):
    print("----------------------Loop - " + str(i) + "----------------------")
    filename = zip_files[i] + ".zip"
    print(f"Found zip file: {filename}")

    # Define the path to the zip file
    zip_path = os.path.join(source_directory, filename)
    print(f"Zip file path: {zip_path}")

    # Define the folder name to extract to (removing the .zip extension)
    extract_folder_name = os.path.splitext(filename)[0]
    extract_path = os.path.join(destination_directory, extract_folder_name)
    print(f"Extraction folder path: {extract_path}")

    # Create the extraction folder if it does not exist
    if not os.path.exists(extract_path):
        os.makedirs(extract_path)
        print(f"Created directory for extraction: {extract_path}")
    else:
        print(f"Directory already exists, skipping creation: {extract_path}")

    # Try to extract the zip file
    try:
        print(f"Starting extraction for {filename}...")
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(extract_path)
            print(f"Extraction complete for {filename} to {extract_path}\n")
        
        # Wait until the file is free to delete
        while True:
            try:
                os.remove(zip_path)
                print(f"Removed: {zip_path}")
                break
            except PermissionError:
                print(f"Waiting for {filename} to be ready for deletion...")
                time.sleep(0.5)  # Check every 0.5 seconds

    except (FileNotFoundError, PermissionError) as e:
        # Log the error and continue to the next file
        print(f"Error handling {filename}: {e}")
        with open(log_file_path, "a") as log_file:
            log_file.write(f"Error handling {filename}: {e}\n")
        continue  # Move to the next file

print("All zip files have been processed.")
```
