```python
from pymongo import MongoClient
import re
import os
from typing import List, Dict, Any
import shutil
```

```python
def connect_mongodb(uri,db_name, collection_name):
    # Connect to MongoDB 
    client = MongoClient(uri)
    
    # Select the database and collection
    db = client[db_name]
    collection = db[collection_name]
    
    return collection
def get_all_directories(root_path):
    """
    Get all directory names under the given root path.
    Returns a list of directory names.
    """
    return [name for name in os.listdir(root_path) if os.path.isdir(os.path.join(root_path, name))]
def contains_original(lst):
    target = "original"
    return any(item.lower() == target for item in lst)
def add_data_to_mongodb(collection, app_id, level_2_path_list):
    # Tạo document
    document = {
        "_id": app_id,
        "level_2_path": level_2_path_list
    }
    try:
        collection.insert_one(document)
        print("Data inserted successfully!")
    except Exception as e:
        print("Error inserting data:", e)
```

```python
mongoDB_uri = 'mongodb://localhost:27017'
mongoDB_database = 'wearable-project-2' 
mongoDB_collection = 'codeblock_ast'
```

```python
# Connect to the MongoDB collection
collection = connect_mongodb(mongoDB_uri,mongoDB_database,mongoDB_collection)
```

```python
root_path = r"E:\wearable-2-dataset-code-extraction"
```

```python
array_root_directory = get_all_directories(root_path)
# print(len(array_directory))
```

```python
for x in range(len(array_root_directory)):
    package_name = array_root_directory[x]
    print("------------ Loop-"+str(x)+" ------------")
    print("package name: ",package_name)
    level_1_path = root_path+"\\"+package_name
    print(level_1_path)
    array_level_1_directory = get_all_directories(level_1_path)
    print(array_level_1_directory)
    check_code_source = contains_original(array_level_1_directory)
    # print(check_code_source)
    level_2_path_array = []
    if(check_code_source):
        level_2_path = level_1_path+"\\original"
        # print(level_2_path)
        level_2_path_array.append(level_2_path)
    else:
        for y in range(len(array_level_1_directory)):
            directory_name_level_2 = array_level_1_directory[y]
            level_2_path = level_1_path+"\\"+directory_name_level_2
            # print(level_2_path)
            level_2_path_array.append(level_2_path)
    print(level_2_path_array)
    add_data_to_mongodb(collection, package_name, level_2_path_array)
    # break
```
