```python
from pymongo import MongoClient
import pandas as pd
```

```python
def connect_mongodb(uri, db_name, collection_name):
    # Connect to MongoDB 
    client = MongoClient(uri)
    
    # Select the database and collection
    db = client[db_name]
    collection = db[collection_name]
    
    return collection
def prepare_documents_from_csv(csv_path):
    df = pd.read_csv(csv_path)

    documents = []
    for _, row in df.iterrows():
        # Split multiple links by newlines
        links = str(row['link']).strip().splitlines()

        document = {
            "id": int(row["id"]),
            "third_party": row["third_party"].strip(),
            "manifest": row["manifest"].strip() if pd.notna(row["manifest"]) else None,
            "source_code": row["source_code"].strip() if pd.notna(row["source_code"]) else None,
            "stop_share": row["stop_share"].strip() if pd.notna(row["stop_share"]) else None,
            "links": links
        }
        documents.append(document)

    return documents
def insert_documents_to_mongodb(collection, documents):
    if documents:
        collection.insert_many(documents)
        print(f"Inserted {len(documents)} documents.")
    else:
        print("No documents to insert.")
```

```python
csv_path = r"C:\Users\ASUS\anaconda3-project-code\wearable-code-extraction\third-party-services.csv"
uri = "mongodb://localhost:27017/"
db_name = "wearable-project-2"
collection_name = "third_party_services"

# Connect to MongoDB
collection = connect_mongodb(uri, db_name, collection_name)

# Prepare documents
documents = prepare_documents_from_csv(csv_path)

# Insert into MongoDB
insert_documents_to_mongodb(collection, documents)
```
