```python
import os
import javalang
from typing import List, Tuple
```

```python
def extract_ast_paths(code: str) -> List[str]:
    try:
        tree = javalang.parse.parse(code)
    except javalang.parser.JavaSyntaxError as e:
        print(f"Error parsing Java code: {e}")
        return []
    
    paths = []
    
    def traverse(node, path):
        if isinstance(node, javalang.ast.Node):
            node_name = type(node).__name__
            path.append(node_name)
            
            for child in node.children:
                if isinstance(child, list):
                    for item in child:
                        traverse(item, path[:])
                else:
                    traverse(child, path[:])
                    
        elif isinstance(node, list):
            for item in node:
                traverse(item, path[:])
        else:
            if node is not None:
                paths.append(".".join(path) + " -> " + str(node))
    
    traverse(tree, [])
    return paths

def read_java_code_from_file(file_path: str) -> str:
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            code = file.read()
        return code
    except FileNotFoundError:
        print(f"File not found: {file_path}")
        return ""
    except Exception as e:
        print(f"Error reading file: {e}")
        return ""
def extract_ast_paths_2(code: str) -> List[str]:
    try:
        tree = javalang.parse.parse(code)
    except javalang.parser.JavaSyntaxError as e:
        print(f"Error parsing Java code: {e}")
        return []
    
    paths = []

    # Các loại node nên ghi nhận
    interested_node_types = (
        javalang.tree.Literal,
        javalang.tree.MemberReference,
        javalang.tree.MethodInvocation,
        javalang.tree.BinaryOperation,
        javalang.tree.ClassCreator,
        javalang.tree.FormalParameter,
        javalang.tree.VariableDeclarator,
    )

    def traverse(node, path):
        if isinstance(node, javalang.ast.Node):
            node_name = type(node).__name__
            path.append(node_name)
            
            # Nếu là node kiểu quan tâm -> lưu lại
            if isinstance(node, interested_node_types):
                value = getattr(node, 'value', None) or getattr(node, 'member', None) or getattr(node, 'name', None)
                if value:
                    paths.append(".".join(path) + " -> " + str(value))
            
            # Duyệt tiếp các con
            for child in node.children:
                if isinstance(child, list):
                    for item in child:
                        traverse(item, path[:])
                elif child is not None:
                    traverse(child, path[:])
                    
        elif isinstance(node, list):
            for item in node:
                traverse(item, path[:])
        else:
            # Không cần lưu những primitive hoặc None ở đây
            pass

    traverse(tree, [])
    return paths
# def extract_code_blocks(code: str) -> List[Tuple[str, str]]:
#     """
#     Trích xuất các code block (method/class/if/for) từ file Java.
#     Trả về danh sách (block_type, block_code).
#     """
#     try:
#         tree = javalang.parse.parse(code)
#     except javalang.parser.JavaSyntaxError as e:
#         print(f"Error parsing Java code: {e}")
#         return []
    
#     blocks = []
#     code_lines = code.splitlines()
    
#     def get_code(start_line, end_line):
#         if start_line is not None and end_line is not None:
#             return "\n".join(code_lines[start_line-1:end_line])
#         return ""

#     for path, node in tree:
#         if isinstance(node, (javalang.tree.MethodDeclaration,
#                               javalang.tree.ClassDeclaration,
#                               javalang.tree.ConstructorDeclaration,
#                               javalang.tree.IfStatement,
#                               javalang.tree.ForStatement,
#                               javalang.tree.WhileStatement,
#                               javalang.tree.SwitchStatement)):
#             block_type = type(node).__name__
            
#             start_line = node.position.line if node.position else None
            
#             # Tìm end_line bằng cách duyệt sâu xuống để lấy node con có dòng số lớn nhất
#             end_line = start_line
#             def find_max_line(n):
#                 nonlocal end_line
#                 if isinstance(n, javalang.ast.Node) and n.position:
#                     if n.position.line > end_line:
#                         end_line = n.position.line
#                     for child in n.children:
#                         if isinstance(child, list):
#                             for item in child:
#                                 find_max_line(item)
#                         elif isinstance(child, javalang.ast.Node):
#                             find_max_line(child)
            
#             find_max_line(node)
            
#             block_code = get_code(start_line, end_line)
#             blocks.append((block_type, block_code))
    
#     return blocks
def extract_small_code_blocks(code: str) -> List[Tuple[str, str]]:
    """
    Trích xuất các code block nhỏ (method, if, for...) từ file Java.
    Không lấy class lớn, chỉ lấy các block riêng biệt nhỏ nhất.
    """
    try:
        tree = javalang.parse.parse(code)
    except javalang.parser.JavaSyntaxError as e:
        print(f"Error parsing Java code: {e}")
        return []
    
    blocks = []
    code_lines = code.splitlines()

    def get_code(start_line, end_line):
        if start_line is not None and end_line is not None:
            return "\n".join(code_lines[start_line-1:end_line])
        return ""

    # Cần lưu các block đã lấy để tránh overlap
    selected_nodes = []

    def find_max_line(n):
        max_line = n.position.line if n.position else -1
        if isinstance(n, javalang.ast.Node):
            for child in n.children:
                if isinstance(child, list):
                    for item in child:
                        if isinstance(item, javalang.ast.Node):
                            max_line = max(max_line, find_max_line(item))
                elif isinstance(child, javalang.ast.Node):
                    max_line = max(max_line, find_max_line(child))
        return max_line

    # Chỉ duyệt các node là method/if/for/while
    for path, node in tree:
        if isinstance(node, (javalang.tree.MethodDeclaration,
                              javalang.tree.ConstructorDeclaration,
                              javalang.tree.IfStatement,
                              javalang.tree.ForStatement,
                              javalang.tree.WhileStatement,
                              javalang.tree.SwitchStatement)):
            # Chỉ lấy node nếu không bị chứa trong node đã chọn trước đó
            is_inside = False
            for other in selected_nodes:
                if node.position and other.position:
                    if other.position.line <= node.position.line:
                        max_line = find_max_line(other)
                        if node.position.line <= max_line:
                            is_inside = True
                            break
            if not is_inside:
                selected_nodes.append(node)
                block_type = type(node).__name__
                start_line = node.position.line if node.position else None
                end_line = find_max_line(node)
                block_code = get_code(start_line, end_line)
                blocks.append((block_type, block_code))
    
    return blocks
```
