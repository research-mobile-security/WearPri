```python
import os
import shutil

def remove_non_java_directories(root_path):
    # Get all directories under the root path that do not end with '-java'
    non_java_dirs = [d for d in os.listdir(root_path) 
                     if os.path.isdir(os.path.join(root_path, d)) and not d.endswith('-java')]
    
    print("Removing directories without '-java' in the name:")
    for directory in non_java_dirs:
        dir_path = os.path.join(root_path, directory)
        try:
            shutil.rmtree(dir_path)
            print(f"Removed directory: {dir_path}")
        except Exception as e:
            print(f"Error removing directory {dir_path}: {e}")

# Define the root path
root_path = r"E:\wearable-project-decompile-source\source-code"

# Run the function
remove_non_java_directories(root_path)
```
