```python
from pymongo import MongoClient
from typing import Dict, Any, Union, List
```

```python
def connect_mongodb(uri,db_name, collection_name):
    # Connect to MongoDB 
    client = MongoClient(uri)
    
    # Select the database and collection
    db = client[db_name]
    collection = db[collection_name]
    
    return collection
def count_privacy_implement_elements(collection):
    # Query to count documents where "manifest" or "source_code" is "yes"
    count = collection.count_documents({
        "$or": [
            {"manifest": "yes"},
            {"source_code": "yes"}
        ]
    })
    return count
def count_stop_share_yes(collection):
    # Query to count documents where "stop_share" is "yes"
    count = collection.count_documents({
        "stop_share": "yes"
    })
    return count
def get_all_third_parties(collection):
    # Query to get all "third_party" fields
    results = collection.find({}, {"third_party": 1, "_id": 0})
    third_parties = [item["third_party"] for item in results]
    return third_parties
def extract_compliance_info(document: Dict[str, Any]) -> Union[Dict[str, Any], None]:
    """
    Extract _id, 3rd_integration, and privacy-compliance.violate if privacy-compliance.privacy-compliance is 'no'.

    Args:
        document (Dict[str, Any]): A single document from the MongoDB collection.

    Returns:
        Dict[str, Any] or None: A dictionary containing the extracted information if conditions are met, otherwise None.
    """
    if document.get("privacy-compliance", {}).get("privacy-compliance") == "no":
        return {
            "_id": document.get("_id"),
            "third-party-keyword": document.get("third-party-keyword"),
            "privacy-compliance.violate": document.get("privacy-compliance", {}).get("violate")
        }
    return None


def fetch_compliance_violations(collection) -> List[Dict[str, Any]]:
    """
    Fetch all documents from the collection and filter those with privacy compliance violations.
    
    Args:
        collection (pymongo.collection.Collection): The MongoDB collection to query.

    Returns:
        List[Dict[str, Any]]: A list of filtered documents with compliance violations.
    """
    results = []
    for document in collection.find():
        filtered_info = extract_compliance_info(document)
        if filtered_info:
            results.append(filtered_info)
    return results
def extract_keywords(data: Dict[str, List[str]]) -> List[str]:
    """
    Extract keywords from the given dictionary and return them as a list.

    Args:
        data (Dict[str, List[str]]): The dictionary containing keywords as keys.

    Returns:
        List[str]: A list of keywords from the dictionary.
    """
    return list(data.keys())
def get_third_parties_with_stop_share(collection):
    # Query to find all documents where "stop_share" is "yes"
    results = collection.find({"stop_share": "yes"}, {"third_party": 1, "_id": 0})
    third_parties = [item["third_party"] for item in results]
    return third_parties
```

```python
# Connect to the MongoDB collection
collection = connect_mongodb(mongoDB_uri,mongoDB_database,mongoDB_collection)
num_privacy_implement = count_privacy_implement_elements(collection)
print("num_privacy_implement:", num_privacy_implement)
```

```python
stop_share = count_stop_share_yes(collection)
print("stop_share: ",stop_share)
```

```python
third_party = get_all_third_parties(collection)
print("third_party: ", third_party)
print("number of third_party: ", len(third_party))
```
