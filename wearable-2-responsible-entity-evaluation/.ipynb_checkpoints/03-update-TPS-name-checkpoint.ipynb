```python
from pymongo import MongoClient, ASCENDING
```

```python
# server
mongoDB_uri = 'mongodb://localhost:27017'
mongoDB_database = 'wearable-project-2' 
mongoDB_collection = 'evaluate-responsible-entity'
```

```python
def connect_mongodb(uri, db_name, collection_name):
    # Connect to MongoDB 
    client = MongoClient(uri)
    
    # Select the database and collection
    db = client[db_name]
    collection = db[collection_name]
    
    return collection
def update_tps_name(old_name="Google", new_name="firebase"):
    """Find all documents with tps_name == old_name and update to new_name."""
    collection = connect_mongodb(mongoDB_uri, mongoDB_database, mongoDB_collection)

    # Build query and update operation
    query = {"tps_name": old_name}
    update = {"$set": {"tps_name": new_name}}

    # Update all matching documents
    result = collection.update_many(query, update)

    print(f"Matched documents: {result.matched_count}")
    print(f"Modified documents: {result.modified_count}")
```

```python
if __name__ == "__main__":
    update_tps_name("Google", "firebase")
```
