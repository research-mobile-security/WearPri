```python
# Function get all sub-directory under root directory
def get_all_subdirectories_under_one_level(root_dir: str):
    """
    Get all subdirectories directly under the given root directory (one level only).
    Return a list of full paths.
    """
    return [
        os.path.join(root_dir, name)
        if not os.path.isabs(name) else name
        for name in os.listdir(root_dir)
        if os.path.isdir(os.path.join(root_dir, name))
    ]
def get_app_name(path: str) -> str:
    """
    Return the last directory name from a full path.
    Example:
        E:\\wearable-2-dataset-code-extraction\\adarshurs.android.vlcmobileremote
        -> 'adarshurs.android.vlcmobileremote'
    """
    return os.path.basename(os.path.normpath(path))
```

```python
FAISS_DIR   = r"C:\Users\ASUS\anaconda3-project-code\wearable-2-sharing-non-compliance-evaluation\faiss_all_apps"  # 1 index chung
INDEX_NAME  = "java_blocks_all"
EMBED_MODEL = "text-embedding-3-small"  # hoặc 'text-embedding-3-large' nếu muốn
```

```python
def _manifest_path() -> Path:
    return Path(FAISS_DIR) / "ingested_manifest.json"

def _load_manifest() -> Dict:
    mp = _manifest_path()
    if mp.exists():
        try:
            return json.loads(mp.read_text(encoding="utf-8"))
        except Exception:
            return {"files": {}}
    return {"files": {}}

def _save_manifest(manifest: Dict):
    Path(FAISS_DIR).mkdir(parents=True, exist_ok=True)
    _manifest_path().write_text(json.dumps(manifest, ensure_ascii=False, indent=2), encoding="utf-8")

def _seen_or_unchanged(manifest: Dict, p: Path) -> bool:
    key = str(p.resolve())
    try:
        mtime = p.stat().st_mtime
    except FileNotFoundError:
        return True
    return manifest["files"].get(key) == mtime

def _mark_ingested(manifest: Dict, p: Path):
    key = str(p.resolve())
    manifest["files"][key] = p.stat().st_mtime

# -------- File helpers --------
def _get_txt_file(dir_path: str) -> Optional[Path]:
    """
    Trả về file .txt duy nhất trong thư mục level_2_path.
    Nếu không tìm thấy, trả về None.
    """
    txt_files = list(Path(dir_path).glob("*.txt"))
    return txt_files[0] if txt_files else None


def _parse_txt_metadata(txt_path: Path) -> Dict:
    """Parse TXT dạng 'key: value' thành dict; tự parse list cho Sink/Source."""
    meta: Dict[str, object] = {
        "app_package_name": None,
        "category": None,
        "rule_name": None,
        "Sink": [],
        "Source": [],
        "Url": None,
        "txt_path": str(txt_path),
    }
    if not txt_path or not txt_path.exists():
        return meta

    for line in txt_path.read_text(encoding="utf-8").splitlines():
        line = line.strip()
        if not line or ":" not in line:
            continue
        key, val = line.split(":", 1)
        key, val = key.strip(), val.strip()
        if key in {"Sink", "Source"}:
            try:
                meta[key] = json.loads(val.replace("'", '"'))
            except Exception:
                inside = re.findall(r"'(.*?)'", val)
                meta[key] = inside if inside else [val]
        else:
            meta[key] = val
    return meta

def _read_java_files(dir_path: str) -> List[Path]:
    return sorted(Path(dir_path).glob("*.java"))

def _chunk_java(java_path: Path, splitter: RecursiveCharacterTextSplitter) -> List[Document]:
    text = java_path.read_text(encoding="utf-8", errors="ignore")
    base_doc = Document(page_content=text, metadata={"source": str(java_path)})
    return splitter.split_documents([base_doc])

def _attach_meta_to_chunks(chunks: List[Document], meta: Dict, subfolder_name: str, app_name: str):
    for ch in chunks:
        ch.metadata.update({
            "doctype": "java_chunk",
            "app_name": app_name,
            "app_package_name": meta.get("app_package_name") or app_name,
            "category": meta.get("category"),
            "rule_name": meta.get("rule_name"),
            "sink_list": meta.get("Sink", []),
            "source_list": meta.get("Source", []),
            "report_url": meta.get("Url"),
            "txt_file": meta.get("txt_path"),
            "subfolder": subfolder_name,  # "0", "1", ...
        })

def _make_txt_doc(meta: Dict, app_name: str) -> Document:
    content = json.dumps({
        "app": app_name,
        "app_package_name": meta.get("app_package_name") or app_name,
        "category": meta.get("category"),
        "rule_name": meta.get("rule_name"),
        "Sink": meta.get("Sink", []),
        "Source": meta.get("Source", []),
        "Url": meta.get("Url"),
    }, ensure_ascii=False, indent=2)
    return Document(
        page_content=content,
        metadata={
            "doctype": "taint_report",
            "app_name": app_name,
            "app_package_name": meta.get("app_package_name") or app_name,
            "category": meta.get("category"),
            "rule_name": meta.get("rule_name"),
            "report_url": meta.get("Url"),
            "txt_file": meta.get("txt_path"),
        },
    )

# --------- API chính: index một level_2_path ---------
def index_one_subfolder_to_faiss(
    level_2_path: str,
    app_name: str,
    api_key: str,
    faiss_dir: str = FAISS_DIR,
    index_name: str = INDEX_NAME,
    embed_model: str = EMBED_MODEL,
    chunk_size: int = 1000,
    chunk_overlap: int = 120,
) -> Dict:
    """
    Index một sub-folder (vd: ...\app.groupcal.www.apk\0) vào FAISS chung.
    - Luôn parse TXT (nếu TXT mới/chỉnh sửa thì thêm doc 'taint_report').
    - Chunk và index các .java (chỉ với file mới/chỉnh sửa).
    - Dùng manifest để incremental append.
    Trả về dict tóm tắt.
    """
    Path(faiss_dir).mkdir(parents=True, exist_ok=True)

    # 1) Manifest
    manifest = _load_manifest()

    # 2) Embedding & splitter
    embeddings = OpenAIEmbeddings(model=embed_model, api_key=api_key)
    splitter   = RecursiveCharacterTextSplitter.from_language(
        language=Language.JAVA, chunk_size=chunk_size, chunk_overlap=chunk_overlap
    )

    # 3) Load hoặc tạo FAISS
    faiss_file = Path(faiss_dir) / f"{index_name}.faiss"
    if faiss_file.exists():
        vectorstore = FAISS.load_local(
            folder_path=faiss_dir,
            index_name=index_name,
            embeddings=embeddings,
            allow_dangerous_deserialization=True,
        )
        existing = True
    else:
        vectorstore = None
        existing = False

    docs_to_add: List[Document] = []
    subfolder_name = os.path.basename(os.path.normpath(level_2_path))

    # 4) TXT
    txt_path = _get_txt_file(level_2_path)
    meta = {}
    if txt_path:
        meta = _parse_txt_metadata(txt_path)
        if not _seen_or_unchanged(manifest, txt_path):
            docs_to_add.append(_make_txt_doc(meta, app_name))
            _mark_ingested(manifest, txt_path)
    else:
        # Không có TXT vẫn cho meta rỗng (nhưng đúng spec bạn nói luôn có TXT)
        meta = {}

    # 5) JAVA
    for jpath in _read_java_files(level_2_path):
        if _seen_or_unchanged(manifest, jpath):
            continue
        try:
            chunks = _chunk_java(jpath, splitter)
            _attach_meta_to_chunks(chunks, meta, subfolder_name, app_name)
            docs_to_add.extend(chunks)
            _mark_ingested(manifest, jpath)
        except Exception as e:
            print(f"❌ Error chunking {jpath}: {e}")

    if not docs_to_add:
        return {"added_docs": 0, "faiss_dir": faiss_dir, "index_name": index_name, "existing_index": existing}

    # 6) Save/append
    if vectorstore is None:
        vectorstore = FAISS.from_documents(docs_to_add, embeddings)
    else:
        vectorstore.add_documents(docs_to_add)

    vectorstore.save_local(faiss_dir, index_name=index_name)
    _save_manifest(manifest)

    return {"added_docs": len(docs_to_add), "faiss_dir": faiss_dir, "index_name": index_name, "existing_index": existing}
```

```python
from typing import Optional, Dict, Any
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS

def classify_code_block_with_faiss(
    code_block: str,
    api_key: str,
    faiss_dir: str,
    index_name: str,
    embed_model: str = "text-embedding-3-small",
    top_k: int = 1,
    score_threshold: Optional[float] = None,  # để đúng yêu cầu bạn có thể để None (không threshold)
    accept_doctypes: Optional[set] = None,    # ví dụ {"java_chunk", "taint_report"}; None = chấp nhận tất cả
) -> Dict[str, Any]:
    """
    Phân loại LEAK / NON-LEAK cho code_block dựa trên vectorDB (FAISS).
    - Lấy top_k=1. Nếu không retrieve ra gì (hoặc fail threshold) => NON-LEAK.
    - Nếu có 1 kết quả => LEAK. Lấy category từ metadata (TXT hoặc chunk).
    Trả về dict gồm verdict, category, rule_name, app_name, source, score, snippet.
    """
    # 1) Embeddings + load FAISS
    embeddings = OpenAIEmbeddings(model=embed_model, api_key=api_key)
    try:
        vs = FAISS.load_local(
            folder_path=faiss_dir,
            index_name=index_name,
            embeddings=embeddings,
            allow_dangerous_deserialization=True,
        )
    except Exception as e:
        # Index chưa có hoặc lỗi => coi như không có kết quả
        return {
            "verdict": "NON-LEAK",
            "reason": f"Cannot load FAISS index: {e}",
            "category": None,
            "rule_name": None,
            "app_name": None,
            "source": None,
            "score": None,
            "snippet": None,
        }

    # 2) Search (lấy score để có thể đặt threshold nếu cần)
    try:
        results = vs.similarity_search_with_score(code_block, k=top_k)
    except Exception as e:
        return {
            "verdict": "NON-LEAK",
            "reason": f"Search error: {e}",
            "category": None,
            "rule_name": None,
            "app_name": None,
            "source": None,
            "score": None,
            "snippet": None,
        }

    if not results:
        # Không retrieve được gì
        return {
            "verdict": "NON-LEAK",
            "reason": "No neighbors returned",
            "category": None,
            "rule_name": None,
            "app_name": None,
            "source": None,
            "score": None,
            "snippet": None,
        }

    # 3) Lấy best hit
    doc, score = results[0]

    # Optional: lọc theo doctype
    if accept_doctypes is not None:
        dt = doc.metadata.get("doctype")
        if dt not in accept_doctypes:
            return {
                "verdict": "NON-LEAK",
                "reason": f"Top result doctype '{dt}' not in {accept_doctypes}",
                "category": None,
                "rule_name": None,
                "app_name": None,
                "source": None,
                "score": score,
                "snippet": (doc.page_content[:200].replace("\n", " ") if doc.page_content else None),
            }

    # Optional: threshold (FAISS score là khoảng cách; càng thấp càng giống).
    if score_threshold is not None and score > score_threshold:
        return {
            "verdict": "NON-LEAK",
            "reason": f"Score {score:.4f} > threshold {score_threshold}",
            "category": None,
            "rule_name": None,
            "app_name": None,
            "source": None,
            "score": score,
            "snippet": (doc.page_content[:200].replace("\n", " ") if doc.page_content else None),
        }

    # 4) Có kết quả → LEAK. Lấy category từ metadata
    md = doc.metadata or {}
    category = md.get("category")
    rule_name = md.get("rule_name")
    app_name  = md.get("app_name")
    source    = md.get("source") or md.get("txt_file")
    snippet   = (doc.page_content[:400].replace("\n", " ") if doc.page_content else None)

    # Chuẩn hóa theo mô tả của bạn: "category: PersonalInfo" => Leak PersonalInfo
    return {
        "verdict": "LEAK",
        "leak_type": category,  # ví dụ: "PersonalInfo"
        "category": category,
        "rule_name": rule_name,
        "app_name": app_name,
        "source": source,
        "score": score,
        "snippet": snippet,
    }
```
