```python
import os
import xml.etree.ElementTree as ET
from datetime import datetime
from pymongo import MongoClient
import json
from typing import Dict, Any, Union, List
from dotenv import load_dotenv
import pandas as pd
```

```python
def connect_mongodb(uri,db_name, collection_name):
    # Connect to MongoDB 
    client = MongoClient(uri)
    
    # Select the database and collection
    db = client[db_name]
    collection = db[collection_name]
    
    return collection
def get_all_ids(collection):
    # Retrieve all _id values
    ids = collection.find({}, {"_id": 1})
    
    # Extract _id values into a list
    id_list = [doc["_id"] for doc in ids]
    
    return id_list
def get_ids_with_privacy_compliance(collection):
    # Query for documents that contain the "privacy-compliance" field
    docs = collection.find({"privacy-compliance": {"$exists": True}}, {"_id": 1})
    
    # Extract and return the list of _id values
    return [doc["_id"] for doc in docs]
def remove_common_elements(list_1, list_2):
    return [item for item in list_1 if item not in list_2]
def get_element_by_id(collection, id_value):
    # Find the document with the given _id
    result = collection.find_one({"_id": id_value})
    return result
def extract_keywords(data: Dict[str, List[str]]) -> List[str]:
    """
    Extract keywords from the given dictionary and return them as a list.

    Args:
        data (Dict[str, List[str]]): The dictionary containing keywords as keys.

    Returns:
        List[str]: A list of keywords from the dictionary.
    """
    return list(data.keys())
def get_third_party_column(csv_path):
    try:
        df = pd.read_csv(csv_path)
        if "third_party" in df.columns:
            return df["third_party"].dropna().tolist()
        else:
            raise KeyError("Column 'third_party' not found in CSV.")
    except Exception as e:
        print(f"Error reading CSV file: {e}")
        return []
def find_not_existing(list_1, list_2):
    # Prepare normalized words from list_1
    keywords = set()
    for item in list_1:
        words = item.strip().lower().split()
        keywords.update(words)  # collect all individual words
    
    not_found = []
    for item in list_2:
        words = item.strip().lower().split()
        if any(word in keywords for word in words):
            continue  # at least one word matched → consider as existing
        not_found.append(item)
    
    return not_found
def append_all(source_list, target_list):
    target_list.extend(source_list)
    return target_list
def remove_duplicates(input_list):
    seen = set()
    result = []
    for item in input_list:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result
```

```python
mongoDB_uri = 'mongodb://localhost:27017'
mongoDB_database_1 = 'wearable-project' 
mongoDB_collection_1 = 'wearable-app-02-04-2025'
```

```python
collection_1 = connect_mongodb(mongoDB_uri, mongoDB_database_1, mongoDB_collection_1)
all_app_id = get_all_ids(collection_1)
print(len(all_app_id))
```

```python
analyzed_app_id = get_ids_with_privacy_compliance(collection_1)
print(len(analyzed_app_id))
```

```python
final_app_id = remove_common_elements(all_app_id, analyzed_app_id)
print(len(final_app_id))
```

```python
final_not_exist = []
for i in range(0,len(final_app_id)):
    print("----------------------- Loop "+str(i)+" -----------------------")
    app_id = final_app_id[i]
    print("App id: ", app_id)
    app_data = get_element_by_id(collection_1,app_id)
    print(app_id)
    third_party_data = app_data["third-party-keyword"]
    third_party_keyword_app = extract_keywords(third_party_data)
    print("third_party_keyword_app: ",third_party_keyword_app)
    third_party_not_exist = find_not_existing(third_party_list, third_party_keyword_app)
    print("third_party_not_exist: ",third_party_not_exist)
    final_not_exist = append_all(third_party_not_exist, final_not_exist)
    # break
# print(final_not_exist)
final_3rd = remove_duplicates(final_not_exist)
```
